```python
import joblib
import pandas as pd
import plotly.express as px
```

```python
clusters = joblib.load('../models/nmf.joblib')
tsne = joblib.load('../models/tsne.joblib')
pca = joblib.load('../models/pca.joblib')

df = pd.read_csv('../data/travel_processed.csv')
df['pred_cluster'] = clusters
df['v1'] = tsne[:,0]
df['v2'] = tsne[:,1]
df["pred_cluster"] = df["pred_cluster"].astype(str)

fig = px.scatter(df, x='v1', y='v2', color='pred_cluster', hover_data=['remarks', 'visit_type'])

fig.update_traces()
fig.show()

df['v1'] = pca[:,0]
df['v2'] = pca[:,1]

fig2 = px.scatter(df, x='v1', y='v2', color='pred_cluster', hover_data=['remarks', 'visit_type'])

fig2.update_traces()
fig2.show()
```
